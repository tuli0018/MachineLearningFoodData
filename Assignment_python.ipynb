```python
modelling_df = foodFacts_3and5Removed[['fat_100g','nutrition_score_uk_100g']].dropna()
modelling_df[:5]
```

```python
modelling_df.plot.scatter(x='fat_100g', y='nutrition_score_uk_100g')
```

```python
mask = np.random.rand(len(modelling_df)) < 0.7
train = modelling_df[mask]
test = modelling_df[~mask]
```

```python
from sklearn.linear_model import LinearRegression
x_train = train[['fat_100g']]
y_train = train['nutrition_score_uk_100g']
model = LinearRegression().fit(x_train,y_train)
```

```python
x_test = test[['fat_100g']]
y_test = test['nutrition_score_uk_100g']
predicted = model.predict(x_test)
model.score(x_test, y_test)
```

```python
from matplotlib import pyplot as plt
plt.scatter(test[['fat_100g']], test['nutrition_score_uk_100g'])
plt.plot(test[['fat_100g']], predicted, color='red')
```

These results do not surprise me since it is clear that items that are high in nutrition majorly have low to medium fat per 100 grams of item.
